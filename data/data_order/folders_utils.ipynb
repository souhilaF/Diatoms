```python
#creer des dossiers de taxons dans le repertoire DATASET_PATH
def creat_folders(DATASET_PATH, DATASET_PATH_DEST):
    files = [f for f in listdir(DATASET_PATH) if os.path.isfile(os.path.join(DATASET_PATH, f))]
    taxons = []
    for file in files:
        taxons.append(file.split('_')[1])
    taxons = set(taxons)
    for taxon in taxons:

        folder = os.path.join(DATASET_PATH_DEST + taxon)
        if not os.path.exists(folder):
            os.makedirs(folder)
        if os.path.exists(folder):
            print('le dossier:' +taxon+ 'existe déjà')
```

```python
#deplace les fichiers de DATASET_PATH_FROM dans DATASET_PATH_DEST
def mouv_file(DATASET_PATH_FROM, DATASET_PATH_DEST):
    files = [f for f in listdir(DATASET_PATH_FROM) if os.path.isfile(os.path.join(DATASET_PATH_FROM, f))]
    
    for file in files:
        print(file)
        os.rename(os.path.join(DATASET_PATH_FROM + file), os.path.join(DATASET_PATH_DEST +'/'+file.split('_')[1]+'/' + file))
    
```
